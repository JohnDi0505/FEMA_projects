```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Hoboken_gcp.csv")
df
```

```python
df2 = df.iloc[:, 0:2]
df2
```

```python
df2.insert(2, "degree_lat", np.nan)
df2.insert(3, "min_lat", np.nan)
df2.insert(4, "sec_lat", np.nan)
df2.insert(5, "degree_lon", np.nan)
df2.insert(6, "min_lon", np.nan)
df2.insert(7, "sec_lon", np.nan)

df2.loc[:, "esrignss_1"] = -df2.loc[:, "esrignss_1"]

df2
```

```python
df2.loc[:, "degree_lat"] = df2.loc[:, "esrignss_l"].astype(int)
df2.loc[:, "min_lat"] = ((df2.loc[:, "esrignss_l"] - df2.loc[:, "degree_lat"]) * 60).astype(int)
df2.loc[:, "sec_lat"] = ((df.loc[:, "esrignss_l"] - df2.loc[:, "degree_lat"] - df2.loc[:, "min_lat"] / 60) * 3600).round(3)

df2.loc[:, "degree_lon"] = df2.loc[:, "esrignss_1"].astype(int)
df2.loc[:, "min_lon"] = ((df2.loc[:, "esrignss_1"] - df2.loc[:, "degree_lon"]) * 60).astype(int)
df2.loc[:, "sec_lon"] = ((df2.loc[:, "esrignss_1"] - df2.loc[:, "degree_lon"] - df2.loc[:, "min_lon"] / 60) * 3600).round(3)

df2
```

```python
df2.iloc[:, 2:].to_csv("Hoboken_gcp.txt", index=None, sep =' ')
```
